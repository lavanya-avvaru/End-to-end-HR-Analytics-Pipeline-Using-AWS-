```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

np.random.seed(42)
random.seed(42)

n = 10000

departments = ['Engineering', 'Sales', 'HR', 'Finance', 'Marketing', 'Operations', 'Legal', 'IT']
locations = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai', 'Pune']
job_levels = ['Junior', 'Mid', 'Senior', 'Lead', 'Manager', 'Director']
education = ['Bachelor', 'Master', 'PhD', 'Diploma']
gender = ['Male', 'Female', 'Non-binary']

dept = np.random.choice(departments, n, p=[0.25,0.20,0.10,0.10,0.10,0.10,0.05,0.10])

salary_map = {
    'Junior': np.random.randint(300000, 600000, n),
    'Mid': np.random.randint(600000, 1000000, n),
    'Senior': np.random.randint(1000000, 1600000, n),
    'Lead': np.random.randint(1600000, 2200000, n),
    'Manager': np.random.randint(2200000, 3000000, n),
    'Director': np.random.randint(3000000, 5000000, n),
}

level = np.random.choice(job_levels, n, p=[0.20,0.25,0.25,0.15,0.10,0.05])
salary = np.array([salary_map[l][i] for i, l in enumerate(level)])

start_date = datetime(2010, 1, 1)
end_date = datetime(2024, 12, 31)
hire_dates = [start_date + timedelta(days=random.randint(0, (end_date - start_date).days)) for _ in range(n)]

performance = np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.40,0.30,0.15])
attrition = np.where((performance <= 2) | (np.random.rand(n) < 0.10), 'Yes', 'No')

df = pd.DataFrame({
    'employee_id': [f'EMP{str(i).zfill(5)}' for i in range(1, n+1)],
    'name': [f'Employee_{i}' for i in range(1, n+1)],
    'age': np.random.randint(22, 60, n),
    'gender': np.random.choice(gender, n, p=[0.55,0.40,0.05]),
    'education': np.random.choice(education, n, p=[0.50,0.35,0.05,0.10]),
    'department': dept,
    'job_level': level,
    'location': np.random.choice(locations, n),
    'salary': salary,
    'hire_date': [d.strftime('%Y-%m-%d') for d in hire_dates],
    'years_at_company': [round((datetime(2024,12,31) - d).days / 365, 1) for d in hire_dates],
    'performance_score': performance,
    'training_hours': np.random.randint(0, 120, n),
    'manager_id': [f'EMP{str(random.randint(1,500)).zfill(5)}' for _ in range(n)],
    'attrition': attrition,
    'last_promotion_year': np.random.randint(2015, 2025, n),
    'satisfaction_score': np.round(np.random.uniform(1.0, 5.0, n), 1),
})

df.to_csv('hr_raw_data.csv', index=False)
print(f"Done! Shape: {df.shape}")
print(df.head())
```
